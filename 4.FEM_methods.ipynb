# FEM methods
This document introduced the three finite differential method schemes applied in the project, which are:
1. Forward-in-time, centered-in-space (FTCS) scheme
2. Laasonen implicit scheme
3. Crank-Nicolson scheme

## FTCS scheme

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.figure(figsize=(6,3))
plt.plot([0,2],[0,0])
plt.plot([1,1],[0,1])
plt.plot([0,1,2,1],[0,0,0,1],'ko',markersize=8)
plt.text(1.1,0.1,'T[n,j]')
plt.text(0.1,0.1,'T[n,j-1]')
plt.text(1.1,1.1,'T[n+1,j]')
plt.text(1.7,0.1,'T[n,j+1]')
plt.xlabel('space')
plt.ylabel('time')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('FTCS explicit scheme',fontsize=12)
plt.axis([-0.5,2.5,-0.5,1.5]);
```

## Laasonen implicit scheme

```python
plt.figure(figsize=(6,3))
plt.plot([0,2],[1,1])
plt.plot([1,1],[0,1])
plt.plot([0,1,2,1],[1,1,1,0],'ko',markersize=8)
plt.text(1.1,0.1,'T[n,j]')
plt.text(0.1,1.05,'T[n+1,j-1]')
plt.text(0.9,1.05,'T[n+1,j]')
plt.text(1.7,1.05,'T[n+1,j+1]')
plt.xlabel('space')
plt.ylabel('time')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('Implicit scheme',fontsize=12)
plt.axis([-0.25,2.5,-0.5,1.5]);
```

## Crank-Nicolson scheme

```python
plt.figure(figsize=(6,3))
plt.plot([0,2],[0,0])
plt.plot([0,2],[1,1])
plt.plot([1,1],[0,1])
plt.plot([0,1,2,0,1,2],[0,0,0,1,1,1],'ko',markersize=8)
plt.text(0.1,0.1,'T[n,j-1]')
plt.text(1.1,0.1,'T[n,j]')
plt.text(1.7,0.1,'T[n,j+1]')
plt.text(0.1,1.1,'T[n+1,j-1]')
plt.text(1,1.1,'T[n+1,j]')
plt.text(1.7,1.1,'T[n+1,j+1]')
plt.xlabel('space')
plt.ylabel('time')
plt.axis('equal')
plt.yticks([0.0,1.0],[])
plt.xticks([0.0,1.0],[])
plt.title('Crank-Nicolson scheme',fontsize=12)
plt.axis([-0.5,2.5,-0.5,1.5]);
```
