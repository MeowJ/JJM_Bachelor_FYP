# Mechanics of options Markets

There are four types of option positions, which are:
1. A long position in a call option
2. A long position in a pull option
3. A short position in a call option
4. A short position in a pull option

The calculation of the payoff and profit of these positions will be explained in this file, before we start the calculation, some ideas involoved should be explained:
1. sT: price of asset at maturity
2. K: strike price
3. payoff: money that the options owner can get from the transaction (without the consideration of premium)
4. profit: money that the options owner can get from the transaction (with the consideration of premium)

Firstly, we set the values of sT, K, and premium:

```python
import numpy as np
import matplotlib.pyplot as plt

# stock price sT
spot_price = 100

# Long call 
strike_price_long_call = 110 #K
premium_long_call = 3.50

# Long put
strike_price_long_put = 90 #K
premium_long_put = 4

# Stock price range at expiration of the option
sT = np.arange(0.5*spot_price,1.5*spot_price,1)
#arange([start,] stop[, step,], dtype=None)
```

Find the payoff and profit for call options by calculation:
1. Long payoff = max(sT-K,0)
2. Long profit = max(sT-k,0)-premium
3. Short payoff = -max(sT-K,0)
4. Short profit = -(max(sT,0)-premium)

```python
def call_profit(sT, strike_price, premium):
    return (np.where(sT > strike_price, sT - strike_price, 0)-premium)
#Profit= Max(S-K,0)-premium
profit_long_call = call_profit(sT, strike_price_long_call, premium_long_call)
profit_short_call = -call_profit(sT, strike_price_long_call, premium_long_call)

def call_payoff(sT, strike_price):
    return (np.where(sT > strike_price, sT - strike_price, 0))
#Payoff= Max(S-K,0)
payoff_long_call = call_payoff(sT, strike_price_long_call)
payoff_short_call = -call_payoff(sT, strike_price_long_call)

# Plot
%matplotlib inline
%config InlineBackend.figure_format='svg'
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,profit_long_call,label='Profit of Long Call',color='g')
ax.plot(sT,payoff_long_call,'--',label='payoff of Long Call',color='g')                                 
ax.plot(sT,profit_short_call,label='Profit of Short Call',color='b')
ax.plot(sT,payoff_short_call,'--',label='Payoff of Short Call',color='b')
plt.xlabel('Stock Price')
plt.ylabel('Profit (or Payoff)')
plt.legend()
plt.show()

```

Find the payoff and profit for put options by calculation:
1. Long payoff = max(K-sT,0)
2. Long profit = max(K-sT,0)-premium
3. Short payoff = max(K-sT,0)
4. Short profit = -(max(K-sT,0)-premium)

```python
def put_profit(sT, strike_price, premium):
    return np.where(sT < strike_price, strike_price - sT, 0)-premium
      
profit_long_put = put_profit(sT, strike_price_long_put, premium_long_put)
profit_short_put = -put_profit(sT, strike_price_long_put, premium_long_put)

def put_payoff(sT, strike_price):
    return (np.where(sT < strike_price, sT - strike_price, 0))
#Payoff= Max(S-K,0)
payoff_long_put = put_payoff(sT, strike_price_long_call)
payoff_short_put = -put_payoff(sT, strike_price_long_call)

# Plot
%matplotlib inline
%config InlineBackend.figure_format='svg'
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,profit_long_put,label='Profit of Long Put',color='g')
ax.plot(sT,payoff_long_put,'--',label='payoff of Long Put',color='g')                                 
ax.plot(sT,profit_short_put,label='Profit of Short Put',color='b')
ax.plot(sT,payoff_short_put,'--',label='Payoff of Short Put',color='b')
plt.xlabel('Stock Price')
plt.ylabel('Profit')
plt.legend()
plt.show()
```
